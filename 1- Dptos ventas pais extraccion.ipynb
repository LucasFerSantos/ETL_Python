Importamos las Librerias

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36"}
```

Codigo para la extraccion de la pagina


```python
deptos = []
```

```python
def venta (page):
    url = f'https://www.argenprop.com/departamento-venta-pais-argentina-pagina-{page}'
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser') 
    departamento = soup.find_all('main')
    webpage = BeautifulSoup(r.text)
    links = webpage.select("div.card__details-box")
    for item in departamento:
            dpto = {
            'direccion': webpage.select("h2.card__address"),
            'ubicacion': webpage.select("p.card__title--primary"),
            '$': webpage.select("span.card__currency"),
            'precio': webpage.select("p.card__price"),
            'expensas': webpage.select("p.card__expenses"),
            'extras': webpage.select("div.card__monetary-values > ul"),
            
            }
            deptos.append(dpto)
    return
```

Bucle for para poder recorrer todas las paginas del sitio y guardarlas en un dataframe

```python
for x in range(1,9748):
    venta(x)
df = pd.DataFrame(deptos)
```
