Le quitamos las etiquetas html a los datos

```python
df1['Direccion'] = df1['Direccion'].replace({'<h2 class="card__address">':'','</h2>':''}, regex=True ) 
```

```python
df1['Ubicacion'] = df1['Ubicacion'].replace({'<p class="card__title--primary">':'','</p>':''}, regex=True ) 
```

```python
df1['Moneda'] = df1['Moneda'].replace({'<span class="card__currency">':'','</span>':''}, regex=True ) 
```

```python
df1['Precio'] = df1['Precio'].replace({'<p class="card__price">':'','<span class="card__currency">USD</span>':'','</p>':''}, regex=True ) 
```

```python
df1['Expensas'] = df1['Expensas'].replace({'<p class="card__expenses" title="':'','&amp;plus;':''}, regex=True )
```

```python
df1['Extras'] = df1['Extras'].replace({'<ul class="card__main-features">':'','<li>':'', '<i class="icono-superficie_cubierta"></i>':'M2','<i class="icono-cantidad_dormitorios"></i>':'Dormitorios','<i class="icono-antiguedad"></i>':'Antiguedad','<i class="icono-cantidad_banos"></i>':'baños'}, regex=True ) 
```

```python
expensas2 = df1['Expensas'].str.split('expensas">', expand=True) 
```
