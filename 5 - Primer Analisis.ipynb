```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('propiedades promedio.csv')
```

```python
df
```

```python
prom_mean = df.groupby(['Provincia','Habitaciones']).mean()
```

```python
prom_mean
```

```python
prom_mean['Precio'].plot(kind='bar', figsize=(10,10), )
```

```python
prom_mean['Metros Cuadrados'].plot(kind='bar', figsize=(10,10), )
```

```python
prom_mean['Expensas'].plot(kind='bar', figsize=(10,10), )
```
